```python
print("AI Movie Recommender")


genre = input("Pick genre - Action, Comedy, Horror: ").lower()
mood = input("Mood - Happy, Sad, Excited: ").lower()

print(" Recommended for you")

if genre == "action" and mood == "excited":
    print("1. John Wick 4")
    print("2. Mad Max: Fury Road")
    print("3. Top Gun: Maverick")

elif genre == "action" and mood == "happy":
    print("1. Guardians of the Galaxy")
    print("2. Spider-Man: No Way Home")
    print("3. Jumanji")

elif genre == "comedy" and mood == "happy":
    print("1. Superbad")
    print("2. The Hangover")
    print("3. Step Brothers")

elif genre == "comedy" and mood == "sad":
    print("1. Paddington 2")
    print("2. School of Rock")
    print("3. Yes Man")

elif genre == "horror" and mood == "excited":
    print("1. Get Out")
    print("2. A Quiet Place")
    print("3. Hereditary")

elif genre == "horror" and mood == "sad":
    print("1. The Babadook")
    print("2. Hereditary")
    print("3. Midsommar")

else:
    print("No match found. Try Action/Happy or Comedy/Sad")
```
